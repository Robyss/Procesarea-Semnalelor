```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = pd.read_csv('train.csv', index_col='ID')
print(x)
```

```python
datetimes = x['Datetime'].values
counts = x['Count'].values
size = len(x)
```

```python
plt.plot(counts)
plt.show()
```

##### a) Care este frecvența de eșantionare a semnalului din Train.csv

```python
# Perioada de eșantionare este de 1 oră
fs = 1 / 3600   # 1/s = Hz
print(fs)
```

##### b) Ce interval de timp acopera esantioanele din fisier?

```python
import datetime

first_date = datetime.datetime.strptime(datetimes[0], '%d-%m-%Y %H:%M')
last_date = datetime.datetime.strptime(datetimes[-1], '%d-%m-%Y %H:%M')

print(last_date - first_date)
```

```python
fs / 2  # Nyquist frequency
```

```python
N = size

fft = np.abs(np.fft.fft(counts) / N)
fft = fft[0:int(N / 2)]

f = fs * np.linspace(0, N / 2, int(N / 2)) / N

plt.plot(f, fft)
plt.show()

```
