```python
mean = 0    # Media
sigma = 1   # Deviația standard
N = 1000    # Numărul de eșantioane

gaussian = np.random.normal(mean, sigma, N)

plt.hist(gaussian, bins=100)
plt.title("Distribuția Gaussiană Unidimensională")
plt.legend([f'μ: {mean}, σ: {sigma}'])``
plt.show()
```

```python
mean = [0, 0]
sigma = [[1, 0], [0, 1]]
N = 1000

x, y = np.random.multivariate_normal(mean, sigma, N).T

plt.scatter(x, y)
plt.title("Distribuția Gaussiană Bidimensională")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Pasul 1: Alegem covarianța (matricea de covarianță)
covariance_matrix = np.array([[1, 0.5], [0.5, 1]])

# Pasul 2: Stabilim intervalul de lucru
interval = np.linspace(-1, 1, 100)

# Pasul 3: Calculăm matricea de covarianță
# Nu este nevoie să facem acest pas explicit în acest exemplu, deoarece am ales matricea de covarianță în mod direct

# Pasul 4: Eșantionăm din distribuția Gaussiană cu matricea de covarianță C și medie 0
samples = np.random.multivariate_normal(mean=[0, 0], cov=covariance_matrix, size=100)

# Pasul 5: Afișăm valorile pentru S = ℝd
plt.scatter(samples[:, 0], samples[:, 1], label='Eșantioane din distribuția Gaussiană')
plt.title('Eșantionare dintr-o distribuție Gaussiană')
plt.xlabel('Dimensiunea 1')
plt.ylabel('Dimensiunea 2')
plt.legend()
plt.show()

# Pasul 6: Calculăm matricea de covarianță pe intervalul dat
x, y = np.meshgrid(interval, interval)
kernel_matrix = np.dot(np.transpose(np.array([x.flatten(), y.flatten()])), np.array([x.flatten(), y.flatten()]))

# Pasul 7: Afișăm rezultatele
plt.imshow(kernel_matrix, extent=(-1, 1, -1, 1), origin='lower', cmap='viridis')
plt.colorbar(label='Valoarea kernel-ului')
plt.title('Matricea de covarianță calculată prin kernel')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
